```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('Heart.csv')
ds.head(10)
```

```python
x = ds['Age']
y = ds['RestBP']

plt.bar(x, y)
plt.xlabel('Age', fontsize = 12)
plt.ylabel('RestBP', fontsize = 12)

plt.title('Age VS RestBP Bar Plot', fontsize = 20)

plt.show()
```

```python
x = ds['Age']
y = ds['RestBP']

plt.scatter(x, y, alpha = 0.5)
plt.xlabel('Age', fontsize = 12)
plt.ylabel('RestBP', fontsize = 12)

plt.title('Age VS RestBP Scatter Plot', fontsize = 20)

plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
plt.pie(ds['RestECG'].value_counts(), labels=ds['RestECG'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution for RestECG Pie Chart')
plt.show()
```

```python
# Select only numeric columns
numeric_ds = ds.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_matrix = numeric_ds.corr()
corr_matrix
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', linewidth=0.5, fmt=".2f")
plt.title("Correlation matrix of Heart dataset")
plt.show()
```
